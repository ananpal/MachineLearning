```python
import numpy as np
import time

class SimpleRefluxAgent:
    def __init__(self,  enviroment):
        self.enviroment = enviroment
        self.location = (0, 0)
        
    def percieve(self):
        return self.enviroment.percieve(self.location)
    
    def act(self, action):
        self.location = self.enviroment.act(self.location, action)
        return self.location
    
    def learn(self, reward):
        self.enviroment.learn(self.location, reward)
        
```
